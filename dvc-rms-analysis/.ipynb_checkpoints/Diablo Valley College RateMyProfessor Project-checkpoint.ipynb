# Exploratory Data Analysis
***

## Continuous Data
For our numerical data, we will investigate pairwise relationships in the dataset.

```python
sns.pairplot(df)
```

It looks like there is a left-skew for ratings, and a normal distribution with difficulty (centered around a rating of 3). Superficially speaking, this is a good sign for DVC as there is more volume with higher ratings and the difficulty is "just right".

```python
fig,axes = plt.subplots(nrows=1, ncols=2, figsize=(12,4), sharey=True)
sns.histplot(data=df,x='ratings',kde=True,ax=axes[0])
sns.histplot(data=df,x='difficulty',kde=True,ax=axes[1])
```

We can futher investigate the relationship between ratings and difficulty. The data indicates an inverse relationship. This makes sense, as "easier" classes or teachers who teach more intuitively tend to make students succeed in those respective classes, thereby prompting a higher rating. 

```python
fig,axes = plt.subplots(nrows=1, ncols=2, figsize=(12,4), sharey=True)
sns.scatterplot(data=df,x='ratings',y='difficulty',ax=axes[0],alpha=0.5)
sns.lineplot(data=df,x='ratings',y='difficulty',ax=axes[1])
```

## Discrete Data
Specificity can be attained by looking into the different ratings in regards to the subject. With a birds-eye view, we can tell that the "essential" subjects like mathematics, english, and science classes have more ratings than the other. This backs up the notion that many people take these classes because of the general education requirements.

We can also see which departments need a second look with their professsors, as some of the distributions tend to be centered towards lower ratings like theatre and anthropology.

```python
g = sns.FacetGrid(df,col="subjects",col_wrap=7,height=1.5)
g = g.map(plt.hist,"ratings",bins=20)
[plt.setp(ax.texts, text="") for ax in g.axes.flat] # remove the original texts
                                                    # important to add this before setting titles
g.set_titles(row_template='{row_name}',col_template = '{col_name}')
```

## Ethics of RateMyProfessors
The activity on this website and the information it contains affects the number of students that sign up for my class. and perhaps the professor's career as a result. There is only so much a professor can do to the information offered online; it's possible for personal or incorrect information to be shared. Even worse, not much would stop them from convincing 20 of their best friends to follow suit. And the end outcome of this kind of widespread action would be a sharp decline in their user ratings on the website. The site's creators, or the people who hold the data and the processes, really have the final say, and the professor has no meaningful remedy.

But someone else has collected the data, and they are the ones in charge of it. The data that a site like ratemyprofessors.com gathers, is information that they could sell or rent to third parties that they could do what they wish to with. The information is not the professors â€” albeit it's about them. As we debate privacy, it's crucial to keep this distinction between who the data is about and who owns it in mind. 

While this is a serious topic, it's also true that unjustified ratings are typically anomalies (e.g. student gets caught cheating, so rates professor critically). The rating of the professor is usually reflective of the course content and how well the professor can teach concepts. With that said, let's get into the overall interpretation of DVC's courses based on the data.
## Overall Interpretation
RMS's mission is to provide a safe forum to share classroom experiences to help fellow students make critical education choices. While it poses some threat to the professor's personal life and career, the website is often reliable when it contains more reviews, as it's easier to find legit reviews.

On the surface level, it seems DVC is doing well in terms of overall ratings and difficulty rankings. However, one may investigate the departments that seem to be "lacking" in those aforementioned metrics. Of course, they must be cautious of those with not many ratings or unjustified ratings should they choose to do so.
